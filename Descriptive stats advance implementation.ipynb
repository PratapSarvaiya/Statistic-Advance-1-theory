```python
import pandas as pd
import numpy as np
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
# chi-square test
data = sns.load_dataset("tips")
```

```python
data
```

```python
data[["sex","smoker"]]
```

```python
# summarisation of data using crosstab
data_table = pd.crosstab(data['sex'], data['smoker'])
```

```python
data_table
```

```python
observed_values = data_table.values
```

```python
observed_values
```

```python
import scipy.stats as stats
expected_values = stats.chi2_contigency(observed_values)
```
